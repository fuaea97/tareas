```python
v1 = pd.read_csv('Vacunas_2_Dosis.csv', sep = ',', header = 0)
```

```python
v1.head()
```

```python
v1_RM =  v1[v1['Region'] == 'Metropolitana']
v1_RM = v1_RM.dropna()
```

```python
v1_RM['TOTAL_VACUNADOS'] = v1_RM.iloc[:, 5:201].sum(axis = 1)
v1_RM.head()
```

```python
v2_RM = v1_RM.sort_values(by = ['Codigo comuna'])
v2_RM = v2_RM.reset_index()
```

```python
Tablafinal = tb1_RM_COVID.drop(['ANO_DEF','FECHA_DEF', 'SEMANA','GLOSA_SEXO','EDAD_TIPO','EDAD_CANT', 'GLOSA_COMUNA_RESIDENCIA','GLOSA_REG_RES','DIAG1','CAPITULO_DIAG1',
                                  'GLOSA_CAPITULO_DIAG1','CODIGO_GRUPO_DIAG1','GLOSA_GRUPO_DIAG1','CODIGO_CATEGORIA_DIAG1','GLOSA_CATEGORIA_DIAG1','CODIGO_SUBCATEGORIA_DIAG1',
                                  'GLOSA_SUBCATEGORIA_DIAG1','DIAG2','CAPITULO_DIAG2','GLOSA_CAPITULO_DIAG2','CODIGO_GRUPO_DIAG2','GLOSA_GRUPO_DIAG2','CODIGO_CATEGORIA_DIAG2',
                                  'GLOSA_CATEGORIA_DIAG2','CODIGO_SUBCATEGORIA_DIAG2','GLOSA_SUBCATEGORIA_DIAG2'], axis = 1)
```

```python
Tablafinal['HDI'] = R1['HDI']
Tablafinal['TOTAL_VACUNADOS'] = v2_RM['TOTAL_VACUNADOS']
```

```python
Tablafinal
```

```python
a = Tablafinal['CANTIDAD_FALLECIDOS']
b = Tablafinal['HDI']

print(np.corrcoef(a, b))

plt.scatter(a, b, marker = 'o') 
plt.xlabel('Fallecidos por comuna')
plt.ylabel('Riqueza')
plt.plot(np.unique(a), np.poly1d(np.polyfit(a, b, 1))(np.unique(a)), 'r--',color='MAGENTA')
plt.show()
```

```python
a = Tablafinal['TOTAL_VACUNADOS']
b = Tablafinal['HDI']

print(np.corrcoef(a, b))

plt.scatter(a,b, marker = 'o') 
plt.xlabel('Vacunados por Comuna')
plt.ylabel('Riqueza')
plt.plot(np.unique(a), np.poly1d(np.polyfit(a, b, 1))(np.unique(a)), 'r--',color='MAGENTA')
plt.show()
```

```python
model = sm.ols(formula = 'CANTIDAD_FALLECIDOS ~TOTAL_VACUNADOS+HDI', data = Tablafinal).fit()
print(model.params)

print(model.summary())
```
