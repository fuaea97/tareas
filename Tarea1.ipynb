```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from datetime import date
from matplotlib.pyplot import figure
```

```python
tabla1 = pd.read_csv('DEFUNCIONES_FUENTE_DEIS_2016_2021_29042021.csv', sep = ';', encoding = 'latin-1')
```

```python
tabla1.columns = ['ANO_DEF','FECHA_DEF','GLOSA_SEXO','EDAD_TIPO','EDAD_CANT','CODIGO_COMUNA_RESIDENCIA','GLOSA_COMUNA_RESIDENCIA','GLOSA_REG_RES','DIAG1','CAPITULO_DIAG1','GLOSA_CAPITULO_DIAG1','CODIGO_GRUPO_DIAG1','GLOSA_GRUPO_DIAG1','CODIGO_CATEGORIA_DIAG1','GLOSA_CATEGORIA_DIAG1','CODIGO_SUBCATEGORIA_DIAG1','GLOSA_SUBCATEGORIA_DIAG1','DIAG2','CAPITULO_DIAG2','GLOSA_CAPITULO_DIAG2','CODIGO_GRUPO_DIAG2','GLOSA_GRUPO_DIAG2','CODIGO_CATEGORIA_DIAG2','GLOSA_CATEGORIA_DIAG2','CODIGO_SUBCATEGORIA_DIAG2','GLOSA_SUBCATEGORIA_DIAG2']
```

```python
tabla1 = tabla1.fillna("")
```

```python
tabla1
```

```python
tabla1['FECHA_DEF'] = pd.to_datetime(tabla1['FECHA_DEF'])
tabla1['SEMANA'] = tabla1['FECHA_DEF'].dt.week
```

```python
tabla1
```

```python
tabla1_2020_2021 = tabla1[tabla1.ANO_DEF.isin([2020,2021])]
tabla1_2020_2021
```

```python
Tabla1total_gripe = tabla1_2020_2021[tabla1_2020_2021.GLOSA_GRUPO_DIAG1.isin(["Influenza [gripe] y neumonía"])]
```

```python
Tabla1total_gripe
```

```python
Tabla1total_COVID19 = tabla1_2020_2021[tabla1_2020_2021.GLOSA_SUBCATEGORIA_DIAG1.isin(['COVID-19 virus identificado', 'COVID-19 virus no identificado'])]
Tabla1total_COVID19
```

```python
Semanal_Casos_Totales = []
Semanal_Casos_Gripe = []
Semanal_Casos_COVID = []
```

```python
for a in tabla1_2020_2021['SEMANA']:
    Semanal_Casos_Totales.append(a);
    
for b in Tabla1total_gripe['SEMANA']:
    Semanal_Casos_Gripe.append(b);
    
for c in Tabla1total_COVID19['SEMANA']:
    Semanal_Casos_COVID.append(c);
```

```python
Intervalos_Casos_Totales = range(min(Semanal_Casos_Totales), max(Semanal_Casos_Totales) )
Intervalos_Casos_Gripe = range(min(Semanal_Casos_Gripe), max(Semanal_Casos_Gripe))
Intervalos_Casos_COVID = range(min(Semanal_Casos_COVID), max(Semanal_Casos_COVID))
```

```python
%matplotlib inline
lista1 = [1,2,3,4,5]
plt.hist(Semanal_Casos_Totales, bins = Intervalos_Casos_Totales,color = 'black', rwidth = 0.7)
plt.hist(Semanal_Casos_COVID, bins = Intervalos_Casos_COVID, color = 'red', rwidth = 0.7)
plt.title('Grafico casos de Fallecidos totales vs Casos de COVID-19 fallecidos')
plt.xlabel('Semanas')
plt.ylabel('Casos')
plt.show()
```

Se observa que los fallecidos totales y los casos de COVID-19 se comportan igual, y ademas se puede observar que los casos de COVID-19 aumentaron demasiado entre la semana 20  provocando un aumento en los fallecidos totales.

```python
plt.hist(Semanal_Casos_COVID, bins = Intervalos_Casos_COVID, color = 'red', rwidth = 0.7)
plt.hist(Semanal_Casos_Gripe, bins = Intervalos_Casos_Gripe, color = 'yellow', rwidth = 0.85)
plt.title('Grafico casos fallecidos por COVID-19 vs Casos fallecidos por gripe')
plt.xlabel('Semanas')
plt.ylabel('Casos')
plt.show()
```

Se observa que la cantidad de casos de muerte COVID-19 es realmente superior a los casos de muerte por gripe, ademas aclarar que los casos de muerte por COVID -19 tiene un aumento extremandamente mayor entre las semanas 20 y 30.
